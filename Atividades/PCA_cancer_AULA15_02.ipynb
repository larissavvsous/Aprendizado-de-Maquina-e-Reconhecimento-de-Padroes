## **Diferença entre StandardScaler e RobustScaler**

#### A diferença principal entre o `StandardScaler` e o `RobustScaler` reside na forma como tratam os dados em relação à sua **distribuição** e à **presença de outliers**:

- `StandardScaler`: Assume que os dados estão **distribuídos normalmente** e dimensiona os recursos de modo que a **média seja 0 e o desvio padrão seja 1**. Isso significa que ele **centraliza** os dados em torno de 0 e redimensiona-os para que a **variância seja a mesma** em todos os recursos.

- `RobustScaler`: É **menos sensível a outliers**. Ele dimensiona os recursos usando estatísticas que são robustas a outliers, como a **mediana** e os **quartis**. Isso o torna uma **escolha mais segura** quando os dados **têm outliers** ou quando a **distribuição não é normal**.



## **Interpretação das Matrizes**

- **Componentes Principais (Principal Components)**: A primeira matriz, chamada `data_comp`, mostra os componentes principais resultantes da análise de PCA. Cada linha dessa matriz representa um componente principal e cada coluna representa um recurso original do conjunto de dados. Os valores dentro da matriz indicam a importância de cada recurso na formação desse componente principal. Recursos com valores absolutos mais altos têm uma influência maior no componente principal correspondente.

- **Matriz de Correlação**: A matriz de correlação entre os recursos originais do conjunto de dados. Cada célula nessa matriz representa o coeficiente de correlação entre os recursos correspondentes. Coeficientes de correlação próximos a 1 indicam uma correlação positiva forte, enquanto coeficientes próximos a -1 indicam uma correlação negativa forte. Coeficientes próximos a 0 indicam uma correlação fraca ou inexistente. Uma alta correlação positiva entre dois recursos significa que eles tendem a aumentar e diminuir juntos, enquanto uma alta correlação negativa significa que um recurso aumenta enquanto o outro diminui.
